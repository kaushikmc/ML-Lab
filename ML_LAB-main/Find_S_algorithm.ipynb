```python
dataset = [
    ['Sunny', 'Warm', 'Normal', 'Strong', 'Warm', 'Same', 'Yes'],
    ['Sunny', 'Warm', 'High', 'Strong', 'Warm', 'Same', 'Yes'],
    ['Rainy', 'Warm', 'High', 'Strong', 'Warm', 'Change', 'No'],
    ['Sunny', 'Warm', 'High', 'Strong', 'Cool', 'Change', 'Yes'],
    ['Rainy', 'Cool', 'High', 'Strong', 'Warm', 'Change', 'No']
]
# Extract features and labels
features = [row[:-1] for row in dataset]
labels = [row[-1] for row in dataset]
def find_s_algorithm(features, labels):
    # Step 1: Initialize hypothesis with the first positive example
    for i in range(len(labels)):
        if labels[i] == "Yes":
            hypothesis = features[i].copy()
            break
    # Step 2: Update hypothesis for every positive example
    for i in range(len(labels)):
        if labels[i] == "Yes":
            for j in range(len(hypothesis)):
                if hypothesis[j] != features[i][j]:
                    hypothesis[j] = "?"

    return hypothesis
# Run Find-S

hypothesis = find_s_algorithm(features, labels)
print("Final S-Maximal Hypothesis:")
print(hypothesis)
```
